## Κωδικοποίηση
> Ένα άτομο θα αναπαριστάται από ένα διάνυσμα μεγέθους 12. Η κάθε τιμή του διανύσματος θα βρίσκεται στο διάστημα $[0,1]$.

## Πλεονάζουσες τιμές
> Υπάρχουν πολλοί τρόποι για να εξασφαλίσουμε ότι δεν πρόκειται να υπάρξουν πλεονάζουσες τιμές.Παρακάτω αναλύω κάποιους από αυτούς. <br><br> 
**Μηχανισμός Επιδιόρθωσης**: Μια μέθοδος είναι να αλλάζουμε τις τιμές των ατόμων έτσι ώστε να μην υπάρχουν τιμές εκτός ορίων. <br><br>
**Συνάρτηση Ποινής**: Μια άλλη προσέγγιση είναι να χρησιμοποιήσουμε την συνάρτηση καταλληλότητας για να διώχνουμε τέτοια άτομα από τον πληθυσμό ορίζοντας κάποια ποινή για τα άτομα που περιέχουν μη επιτρεπτές τιμές στο διάνυσμά τους.<br><br>
**Φιλτράρισμα**: Μια ακόμα ιδέα είναι να απορρίπτουμε τα ελατωματικά άτομα και να τα αντικαταστούμε με αποδεκτά.<br><br>
**Τι θα χρησιμοποιήσω**: Στο πρότζεκτ επιλέγω να χρησιμοποιήσω την συνάρτηση ποινής καθώς θεωρώ ότι είναι πιο εύκολα υλοποιήσιμη τεχνική καθώς το φιλτράρισμα το κάνει ο ίδιος ο αλγόριθμος.

**Τι θα χρησιμοποιήσω**: Στο προτζεκτ καθώς δεν έχω κάποια έξτρα γνώση για το ποιά άτομα θα ήταν πιο ευνοικά , θα χρησιμοποιήσω το Random Sampling.


## Υπολογισμός ομοιότητας
> Για τον υπολογισμό της ομοιότητας ενδείκνεται να χρησιμοποιήσουμε την ομοιότητα συνιμητόνου. Αυτή είναι μία πρακτική που χρησιμοποιείται σε προβλήματα , όπου είναι αναγκαία η ποσοτικοποίηση της ομοιότητας δύο διανυσμάτων. Η συνάρτηση αυτή παίρνει τιμές στο διάνυσμα $[-1,1]$ (το -1 σημαίνει ότι τα διανύσματα είναι τελείως διαφορετικά και το 1 σημαίνει ότι είναι ακριβώς τα ίδια).
## Συνάρτηση καταλληλότητας
> Για να κατασκευάσουμε την συνάρτηση καταλληλότητας θα χρησιμοποιήσουμε δύο κανόνες:<br>
Ένα άτομο είναι πιο κατάλληλο , εάν 
1. Είναι πιο κοντά στις μετρήσεις των αισθητήρων για τη συγκεκριμένη
κατάσταση/στάση.
2. Απέχει περισσότερο από τις μετρήσεις των αισθητήρων που αντιστοιχούν στις
υπόλοιπες καταστάσεις.<br>

>Επομένως χρησιμοποιώντας την ομοιότητα συνιμοτόνου όπως αναφέρθηκε παραπάνω η F θα μπορούσε να είναι η εξής:<br>
$F(v) = \frac{cos(v,t_j) + c(1 - \frac{1}{4}Σ_{i!=j}cos(v,t_i))}{1+c}$<br>
Όπου $v$ είναι το διάνυσμα του ατόμου, $t_j$ είναι ο ΜΟ των ανθρώπινων μετρήσεων για την κατάσταση j και $c$ είναι μία σταθερά στο διάστημα $[0,1]$<br>

> Η F παίρνει τιμές στο διάστημα $[0,1]$ καθώς είναι συνεχής και η μέγιστη τιμή είναι 1 (Επειδή τα διανύσματα των ατόμων είναι μη αρνητικά τότε η cos παίρνει τιμές από 0 έως 1. Επομένως ο πρώτος όρος του αριθμητή έχει μέγιστη τιμή 1 και ο δεύτερος έχει μέγιστη τιμή c. Επομένως εάν κάνουμε την διαίρεση το αποτέλεσμα είναι 1) και η ελάχιστη είναι 0 (Βγαίνει με την ίδια λογική με παραπάνω)<br>

> Ο παραπάνω τύπος είναι κατάλληλος εάν θεωρήσουμε το πρόβλημα μας ώς ένα πρόβλημα μεγιστοποίησης.<br>

> Μια έγκυρη τιμή που θα μπορούσαμε να θέσουμε στο c είναι το 0.3 έτσι ώστε να μην κυριαρχεί κατά την αξιολόγηση ενός ατόμου αλλά και να μην είναι αμελητέα η επίδρασή του.<br>

> Η συνάρτηση αξιολόγησης $F$ θα παίρνει μια πολύ χαμηλή τιμή (πχ 0) για πλεονάζουσες τιμές.  

## Γενετικοί Τελεστές

Οι γενετικοί τελεστές χωρίζονται σε τρείς κατηγορίες:

> **Τελεστές επιλογής**: <br> 
1. **Επιλογή με χρήση ρουλέτας με βάση το κόστος**. Σε αυτή την μέθοδο , θέτουμε μια πιθανότητα σε κάθε άτομο του πληθυσμού με βάση την συνάρτηση καταλληλότητας. Άτομα τα οποία έχουν μεγαλύτερο σκορ καταλληλότητας , έχουν μεγαλύτερη πιθανότητα να επιλεχθούν. 
2. **Επιλογή με βάση την κατάταξη**. Σε αυτή την μέθοδο , τα άτομα κατατάσσονται με βάση την συνάρτηση καταλληλότητας και ύστερα τους αναθέτεται μια πιθανότητα με βάση την κατάταξη. Τα άτομα που βρίσκονται υψιλότερα στην κατάταξη έχουν μεγαλύτερη πιθανότητα να επιλεχθούν.
3. **Επιλογή με χρήση τουρνουά**. Σε αυτή την μέθοδο , διαλέγουμε ένα υποσύνολο και ύστερα επιλέγουμε το καλύτερο άτομο σε αυτό για αναπαραγωγή.

> **Τελεστές Διασταύρωσης**: <br>
1. **Διασταύρωση μονού σημείου**: Σε αυτή την μέθοδο , διαλέγουμε ένα σημείο στο διάνυσμα (χρωμόσωμα) των γονέων και ενώνουμε το γενετικό περιεχόμενο πριν το σημείο του ενός γονέα με το γενετικό περιεχόμενο μετά το σημείο του άλλου γονέα.
2. **Διασταύρωση πολλαπλού σημείου**: Αυτή η μέθοδος είναι η γενικευμένη μορφή της παραπάνω. Σε αυτή την πρακτική , διαλέγουμε πολλαπλά σημεία και ανταλάσσουμε τα γενετικά περιεχόμενα των γονέων ανά δύο διαστήματα που δημιουργούν τα σημεία.
3. **Ομοιόμορφη Διασταύρωση**: Σε αυτή την μέθοδο , διαλέγουμε τυχαία κάθε bit από τα χρωμοσώματα των γονέων για να φτιάξουμε ένα παιδί. 

> **Τελεστές Μετάλλαξης**: <br>
1. **Ελιτισμός**: Μια πρακτική παραλλαγή της γενικής διαδικασίας της κατασκευής ενός νέου πληθυσμού είναι να επιτρέψει στον καλύτερο άτομο από τη σημερινή γενιά να μεταφέρθει στο επόμενο, αναλλοίωτο. Η στρατηγική αυτή είναι γνωστή ως ελιτιστική επιλογή και εγγυάται ότι η ποιότητα που λαμβάνεται από τον γενετικό αλγόρυθμο, δεν θα μειωθεί από τη μια γενιά στην επόμενη.
2. **Τυχαία Μετάλλαξη**: Σε αυτή την μέθοδο , επιλέγονται τυχαία κάποια bit του χρωμοσώματος και τους ανατίθεται μία τυχαία τιμή.
3. **Μετάλλαξη Φλίπας**: Σε αυτή την μέθοδο , ένα τυχαίο bit του χρωμοσώματος αλλάζει από 0 σε 1 ή το αντίθετο.
4. **Gaussian Μετάλλαξη**: Είναι παρόμοια με την τυχαία μετάλλαξη με την διαφορά ότι οι τιμές ανάθεσης επιλέγονται από μία gaussian κατανομή με μηδενική μέση τιμή και επιλεγμένη διασπορά.

**Συμπέρασμα**: Μετά από μελέτη των μεθόδων κατέλληξα ότι θα χρησιμοποιήσω την επιλογή με χρήση ρουλέτας, διασταύρωηση μονού σημείου καθώς το διάνυσμα που αναπαριστά ένα άτομο είναι μικρό , τυχαία μετάλλαξη καθώς το χρωμόσωμα αποτελείται από δεκαδικές τιμές σε ένα συγκεκριμένο διάστημα και ελιτισμό επειδή ξέρω ότι η λύση είναι μοναδική οπότε θέλω να κρατήσω στον πληθυσμό το καλύτερο άτομο για πιο γρήγορη σύγκλιση.

### Υπολογισμός Κανονικοποιημένων Μέσων Όρων





```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# read dataset #
dataset = pd.read_csv("dataset-HAR-PUC-Rio.csv",delimiter=";", low_memory=False)

# create the scaler #
scaler = MinMaxScaler()

selected_dataset = dataset[['x1','y1','z1','x2','y2','z2','x3','y3','z3','x4','y4','z4']]

normalized_selected_dataset = scaler.fit_transform(selected_dataset)

# initialize mean array
mean_values = {}

for classes in dataset['class'].unique():
  # get subspace
  selected_dataset = dataset[dataset['class'] == classes]

  # get columns
  selected_columns = selected_dataset[['x1','y1','z1','x2','y2','z2','x3','y3','z3','x4','y4','z4']]

  # normalize values
  normalized_selected_columns = scaler.fit_transform(selected_columns)
  
  # find mean
  mean_values[classes] = normalized_selected_columns.mean(axis=0)
```

### Κώδικας Γενετικού Αλγορίθμου


```python
import random
from deap import base, creator, tools
from numpy import dot
from numpy.linalg import norm


# Define the problem parameters
INDIVIDUAL_SIZE = 12
POPULATION_SIZE = [20, 20, 20, 20, 20, 200, 200, 200, 200, 200]
GENERATIONS = 1000
CROSSOVER_PROBABILITY = [0.6, 0.6, 0.6, 0.9, 0.1, 0.6, 0.6, 0.6, 0.9, 0.1]
MUTATION_PROBABILITY = [0.0, 0.01, 0.1, 0.01, 0.01, 0.0, 0.01, 0.1, 0.01, 0.01]
c = 0.3
CLASS = 'sitting'

# Create  ftheitness function and individual type
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)


# Create the toolbox and register the functions
toolbox = base.Toolbox()

toolbox.register("attribute", random.random)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attribute, n=INDIVIDUAL_SIZE)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

def cosine_similarity(a,b):
  return dot(a, b)/(norm(a)*norm(b)) 

# Define the fitness evaluation function
def evaluate(individual):
    return (cosine_similarity(individual, mean_values[CLASS]) + c*(1 - sum(cosine_similarity(individual,value) for key, value in mean_values.items() if key != CLASS))/(1+c)),

# Define the random mutate function
def mutate(individual, mutation_probability):
    for i in range(len(individual)):
        if random.random() < mutation_probability:
            individual[i] = random.random()
    return individual

toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxOnePoint)
toolbox.register("mutate", mutate)
toolbox.register("select", tools.selRoulette)


history = pd.DataFrame({})

for i in range(len(POPULATION_SIZE)):
  print(i)
  
  # Initialize history
  best_individuals_list = pd.DataFrame({})
  for j in range(10):

    # Initialize history
    best_individuals = []

    # Create the population and evolve it
    population = toolbox.population(n=POPULATION_SIZE[i])

    fitnesses = list(map(toolbox.evaluate, population))
    for ind, fit in zip(population, fitnesses):
          ind.fitness.values = fit

    Generation_Num = 0
    
    while Generation_Num < 1000: 

        Generation_Num += 1 

        # Select the next generation individuals
        offspring = tools.selBest(population,k=1)+ toolbox.select(population, k=len(population)-1)
        # Clone the selected individuals to avoid unwanted interactions between them
        offspring = list(map(toolbox.clone, offspring))
        
        for child1, child2 in zip(offspring[::2], offspring[1::2]):
            if random.random() < CROSSOVER_PROBABILITY[i]:
                toolbox.mate(child1, child2)
                del child1.fitness.values
                del child2.fitness.values
        
        for mutant in offspring:
            if random.random() < MUTATION_PROBABILITY[i]:
                toolbox.mutate(mutant,MUTATION_PROBABILITY[i])
                del mutant.fitness.values
        
        # Evaluate the individuals with an invalid fitness
        fresh_individuals = [ind for ind in offspring if not ind.fitness.valid]
        fitnesses = toolbox.map(toolbox.evaluate, fresh_individuals)
        for ind, fit in zip(fresh_individuals, fitnesses):
            ind.fitness.values = fit
        
        population[:] = offspring

        best_individuals.append(tools.selBest(population, k=1)[0].fitness.values[0])    
    best_individuals_list= pd.concat([best_individuals_list, pd.DataFrame({str(j) : best_individuals})], axis=1)
  history = pd.concat([history, pd.Series(best_individuals_list.mean(axis=1), name=str(i))], axis=1)
```

```python
import matplotlib.pyplot as plt
from google.colab import widgets
from google.colab import output


tb = widgets.TabBar([str(i+1) for i in range(len(POPULATION_SIZE))])
for i in range(len(POPULATION_SIZE)):
  with tb.output_to(i, select=(i < len(POPULATION_SIZE))):  

    ax = history.plot(y=str(i), kind='line')
    ax.set_title(f'Population Size = {POPULATION_SIZE[i]} , Crossover Propability = {CROSSOVER_PROBABILITY[i]}, Mutation_Propability = {MUTATION_PROBABILITY[i]}')
    plt.show()
```

> **Population Size**: Συγκρίνοντας το 1 με 6, 2 με 7 κ.λ.π, παρατηρούμε ότι το μέγεθος του πληθυσμού παίζει μεγάλο ρόλο στη μέση καλύτερη λύση. Παρόλα αυτά όσο μεγαλύτερο είναι , τόσο πιο αργός είναι ο αλγόριθμος.

> **Crossover Probability**: Συγρκίνοντας το 2 με 4 και 5, 7 με 9 και 10 , παρατηρούμε ότι η τιμή της πιθανότητας διασταύρωσης δεν παίζει σχετικά μεγάλο ρόλο στην τιμή της καλύτερης λύσης. Παρόλα αυτά καθορίζει τον ρυθμό σύγκλισης του αλγορίθμου. Μεγαλύτερες τιμές της σταθεράς αυτής , προβλέπουν γρηγορότερη σύγκλιση σύμφωνα με τα δεδομένα.

> **Mutation Probability**: Συγκρίνοντας το 1 με 2 και 3 , 6 με 7 και 8 , παρατηρούμε ότι η πιθανότητα μετάλλαξης επιρρεάζει και την τιμή της καλύτερης λύσης αλλά και τον ρυθμό σύγκλισης του αλγορίθμου. Παρατηρείται ότι οι τιμές που ενδείκνυται είναι σχετικά μικρές.
