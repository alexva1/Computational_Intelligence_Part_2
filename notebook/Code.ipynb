## Κωδικοποίηση
> Ένα άτομο θα αναπαριστάται από ένα διάνυσμα μεγέθους 12. Η κάθε τιμή του διανύσματος θα βρίσκεται στο διάστημα $[0,1]$.

## Πλεονάζουσες τιμές
> Υπάρχουν πολλοί τρόποι για να εξασφαλίσουμε ότι δεν πρόκειται να υπάρξουν πλεονάζουσες τιμές.Παρακάτω αναλύω κάποιους από αυτούς. <br><br> 
**Μηχανισμός Επιδιόρθωσης**: Μια μέθοδος είναι να αλλάζουμε τις τιμές των ατόμων έτσι ώστε να μην υπάρχουν τιμές εκτός ορίων. <br><br>
**Συνάρτηση Ποινής**: Μια άλλη προσέγγιση είναι να χρησιμοποιήσουμε την συνάρτηση καταλληλότητας για να διώχνουμε τέτοια άτομα από τον πληθυσμό ορίζοντας κάποια ποινή για τα άτομα που περιέχουν μη επιτρεπτές τιμές στο διάνυσμά τους.<br><br>
**Φιλτράρισμα**: Μια ακόμα ιδέα είναι να απορρίπτουμε τα ελατωματικά άτομα και να τα αντικαταστούμε με αποδεκτά.<br><br>
**Τι θα χρησιμοποιήσω**: Στο πρότζεκτ επιλέγω να χρησιμοποιήσω την συνάρτηση ποινής καθώς θεωρώ ότι είναι πιο εύκολα υλοποιήσιμη τεχνική καθώς το φιλτράρισμα το κάνει ο ίδιος ο αλγόριθμος.

## Αρχικός πληθυσμός
Υπάρχουν πολλές μεθόδοι για να διαλέξουμε τον αρχικό πληθυσμό.<br><br>
**Random Sampling**: Μια από τις πιο απλές και σύνηθες είναι να επιλέξουμε τα άτομα τυχαία.<br><br>
**Heuristic Sampling**: Εάν έχουμε κάποια παραπάνω γνώση για το πρόβλημα μπορούμε να την χρησιμοποιήσουμε για να δημιουργήσουμε τον αρχικό πληθυσμό.<br><br>
**Τι θα χρησιμοποιήσω**: Στο προτζεκτ καθώς δεν έχω κάποια έξτρα γνώση για το ποιά άτομα θα ήταν πιο ευνοικά , θα χρησιμοποιήσω το Random Sampling.


## Υπολογισμός ομοιότητας
> Για τον υπολογισμό της ομοιότητας ενδείκνεται να χρησιμοποιήσουμε την ομοιότητα συνιμητόνου. Αυτή είναι μία πρακτική που χρησιμοποιείται σε προβλήματα , όπου είναι αναγκαία η ποσοτικοποίηση της ομοιότητας δύο διανυσμάτων. Η συνάρτηση αυτή παίρνει τιμές στο διάνυσμα $[-1,1]$ (το -1 σημαίνει ότι τα διανύσματα είναι τελείως διαφορετικά και το 1 σημαίνει ότι είναι ακριβώς τα ίδια).
## Συνάρτηση καταλληλότητας
> Για να κατασκευάσουμε την συνάρτηση καταλληλότητας θα χρησιμοποιήσουμε δύο κανόνες:<br>
Ένα άτομο είναι πιο κατάλληλο , εάν 
1. Είναι πιο κοντά στις μετρήσεις των αισθητήρων για τη συγκεκριμένη
κατάσταση/στάση.
2. Απέχει περισσότερο από τις μετρήσεις των αισθητήρων που αντιστοιχούν στις
υπόλοιπες καταστάσεις.<br>

>Επομένως χρησιμοποιώντας την ομοιότητα συνιμοτόνου όπως αναφέρθηκε παραπάνω η F θα μπορούσε να είναι η εξής:<br>
$F(v) = \frac{cos(v,t_j) + c(1 - \frac{1}{4}Σ_{i!=j}cos(v,t_i))}{1+c}$<br>
Όπου $v$ είναι το διάνυσμα του ατόμου, $t_j$ είναι ο ΜΟ των ανθρώπινων μετρήσεων για την κατάσταση j και $c$ είναι μία σταθερά στο διάστημα $[0,1]$<br>

> Η F παίρνει τιμές στο διάστημα $[0,1]$ καθώς είναι συνεχής και η μέγιστη τιμή είναι 1 (Επειδή τα διανύσματα των ατόμων είναι μη αρνητικά τότε η cos παίρνει τιμές από 0 έως 1. Επομένως ο πρώτος όρος του αριθμητή έχει μέγιστη τιμή 1 και ο δεύτερος έχει μέγιστη τιμή c. Επομένως εάν κάνουμε την διαίρεση το αποτέλεσμα είναι 1) και η ελάχιστη είναι 0 (Βγαίνει με την ίδια λογική με παραπάνω)<br>

> Ο παραπάνω τύπος είναι κατάλληλος εάν θεωρήσουμε το πρόβλημα μας ώς ένα πρόβλημα μεγιστοποίησης.<br>

> Μια έγκυρη τιμή που θα μπορούσαμε να θέσουμε στο c είναι το 0.3 έτσι ώστε να μην κυριαρχεί κατά την αξιολόγηση ενός ατόμου αλλά και να μην είναι αμελητέα η επίδρασή του.<br>

> Η συνάρτηση αξιολόγησης $F$ θα παίρνει μια πολύ χαμηλή τιμή (πχ 0) για πλεονάζουσες τιμές.
